```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

```python
print(arr[0])       # first element
print(arr[1:4])     # slicing
```

```python
print(arr + 10)
print(arr * 2)
```

```python
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
```

```python
arr2 = np.array([1,2,3,4,5,6])
print(arr2.reshape(2,3))
```

```python
np.save("data.npy", arr)
loaded = np.load("data.npy")
print(loaded)
```
